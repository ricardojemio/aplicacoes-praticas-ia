```python
# Instalação da biblioteca oficial da Ultralytics
!pip install ultralytics -q

# Importação bibliotecas
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Inicio e carregamento dos modelos
modelo_nano = YOLO('yolov8n.pt')
modelo_small = YOLO('yolov8s.pt')
modelo_medium = YOLO('yolov8m.pt')

# teste de exemplo
imagem_caminho = 'imagem_teste.jpg'

print("--- Executando Inferência com YOLOv8n ---")
resultados_nano = modelo_nano.predict(source=imagem_caminho, save=True, conf=0.25)

print("--- Executando Inferência com YOLOv8s ---")
resultados_small = modelo_small.predict(source=imagem_caminho, save=True, conf=0.25)

print("--- Executando Inferência com YOLOv8m ---")
resultados_medium = modelo_medium.predict(source=imagem_caminho, save=True, conf=0.25)
```
